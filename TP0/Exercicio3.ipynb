### 3. a
Criar corpos finitos primos, sendo p = 31, 127, 8191 e 131071.

```python
p1 = 31
fp1 = GF(p1)
print(fp1)

p2 = 127
fp2 = GF(p2)
print(fp2)

p3 = 8191
fp3 = GF(p3)
print(fp3)

p4 = 131071
fp4 = GF(p4)
print(fp4)
```

### 3. b

### 3. b
Determinar um elemento primitivo de cada um dos corpos e verificar a proposição

```python
pe1 = fp1.primitive_element()
print("Primitive element of {} is {}".format(p1,pe1))
n = mod(0,p1-1)
pe1^n == 1
```

```python
pe2 = fp2.primitive_element()
print("Primitive element of {} is {}".format(p2,pe2))
n = mod(0,p2-1)
pe2^n == 1
```

```python
pe3 = fp3.primitive_element()
print("Primitive element of {} is {}".format(p3,pe3))
n = mod(0,p3-1)
pe3^n == 1
```

```python
pe4 = fp4.primitive_element()
print("Primitive element of {} is {}".format(p4,pe4))
n = mod(0,p4-1)
pe4^n == 1
```
